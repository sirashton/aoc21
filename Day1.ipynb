```python
import pandas as pd
import numpy as np
```

# Part 1 - Up or Down?
---

```python
data = pd.read_csv('Data_Day1A.txt', sep='\n', header = 0)
data['Shift'] = data['Original'].shift(1)
data['Change'] = np.sign(data['Original'] - data['Shift'])
data['Change'] = data['Change'].map({-1:'Down',1:'Up'})
data
```

```python
data['Change'].value_counts()
```

# Part 2 - Smooth up or down
---

```python
smooth = pd.DataFrame()
smooth['Smooth'] = np.convolve(data['Original'],[1,1,1], 'valid')
smooth['SmoothShift'] = smooth['Smooth'].shift(1)
smooth['Change'] = np.sign(smooth['Smooth'] - smooth['SmoothShift'])
smooth['Change'] = smooth['Change'].map({-1:'Down',1:'Up'})
smooth
```

```python
smooth['Change'].value_counts()
```
